```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from glob import glob
import matplotlib.pyplot as plt
```

```python
paths = glob('./csv_data/*.csv')
df = pd.read_csv(paths[0], index_col = 0)
data = df.values
plt.figure()
plt.imshow(data, cmap="magma_r")
plt.colorbar()
plt.show()
```

```python
size = data.shape[0]

xyzdata = []
for i, xdata in enumerate(data):
    for j, ydata in enumerate(xdata):
        xyzdata.append([i - (size-1) / 2, j - (size-1) / 2, ydata * 50])
```

```python
pred = KMeans(n_clusters=6).fit(xyzdata)
clusted = np.array(pred.labels_)
clusted.shape = [15,15]
plt.figure()
plt.imshow(clusted, cmap="magma_r")
plt.colorbar()
plt.show()
```

```python
labels = pred.labels_
cc = pred.cluster_centers_[:,2:]
center = []
for i, c in enumerate(labels):
    center.append(int(cc[int(c)]))
    
center = np.array(center)
center.shape = [15,15]
plt.figure()
plt.imshow(center, cmap="magma_r")
plt.colorbar()
plt.show()
```
